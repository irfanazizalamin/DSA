```python
print('#Soal 2 \n')
print('Ya, karena besar rentangan untuk tiap attribute berbeda-beda')
print('Maka setiap attribute harus disesuaikan skalanya')

import numpy as np

# menghilangkan atribute yang tidak ikut dalam perhitungan
birds_without_id_labels = birds_clean.drop(columns=['id', 'type'])
data = birds_without_id_labels

normalized_data=((data-data.min())/(data.max()-data.min()))

print(normalized_data)
```

```python
print('#Soal 3 \n')

# mean
mean_vec = np.mean(normalized_data, axis=0)
# print(mean_vec)

# attribute dari dataframe
attribute = []
for col in data.columns: 
    attribute.append(col)
    
# meng-adjust data dengan melakukan centering
centered_data = normalized_data.copy()
for i in range(0, 10):
    centered_data[attribute[i]] = data[attribute[i]] - mean_vec[i]
    
# covariance matrix
cov_matrix = np.cov(centered_data.T)
# print(cov_matrix)

# nilai eigen dan vector eigen 
eig_vals, eig_vecs = np.linalg.eig(cov_matrix)

print('a. Eigenvectors: \n%s' %eig_vecs)
print('\nb. Eigenvalues: \n%s' %eig_vals)
```

```python
# Make a list of (eigenvalue, eigenvector) tuples
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]

# Sort the (eigenvalue, eigenvector) tuples from high to low
eig_pairs.sort()
eig_pairs.reverse()

# Visually confirm that the list is correctly sorted by decreasing eigenvalues
print('Eigenvalues in descending order:')
for i in eig_pairs:
    print(i[0])
```

```python
tot = 0
for eig_val in eig_vals:
    tot += eig_val
    
var_exp = [(i / tot) for i in sorted(np.abs(eig_vals), reverse=True)]
cum_var_exp = np.cumsum(var_exp)
plt.bar(range(1, eig_vals.size + 1), var_exp, alpha=0.5, align='center',
        label='Individual')
plt.step(range(1, eig_vals.size + 1), cum_var_exp, where='mid',
         label='Cumulative')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.legend(loc='best')
plt.tight_layout()
plt.savefig('./fig-pca-var-exp.png', dpi=300)
plt.show()

# melihat detail nilai dari tiap basis
for index, val in enumerate(cum_var_exp):
    print('basis ' + str(index+1) +  ' dengan alpha ' + str(val))
```
