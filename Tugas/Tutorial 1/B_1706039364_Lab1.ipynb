```python
import numpy as np
```

```python
risuto = [8, 777, 6, 4, 114514, 7, 21, 42, 888]
```

```python
numpy_array = np.array(risuto)
print('ini adalah list biasa :', risuto, type(risuto))
print('ini adalah numpy array :', numpy_array, type(numpy_array))
```

```python
# mengakses element dengan index 3 -> index mulai dari 0
print('elemen ke-3 :', numpy_array[3])
# menambahkan elemen ke dalam list, dan mengembalikan list tersebut
appended_array = np.append(numpy_array, 9999)
print('original array :', numpy_array)
print('appended array :', appended_array)
```

```python
print('rata-rata dari array:', numpy_array.mean())
print('nilai maksimum dari array:', numpy_array.max())
```

```python
##################  SOAL 1  ##################
print('##################  SOAL 1  ##################')

print('nilai median dari array:', np.median(numpy_array))
print('nilai minimum dari array:', np.min(numpy_array))
print('nilai standar deviasi dari array:', np.std(numpy_array))
print('nilai variance dari array', np.var(numpy_array))
```

```python
sorted_array = np.sort(numpy_array)
print(sorted_array)
```

```python
a = np.array([1, 6, 21])
b = np.array([0.1, 0.2, 0.3])
print('hasil penjumlahan:', np.add(a, b))
print('hasil perkalian:', np.multiply(a, b))
```

```python
##################  SOAL 2  ##################
print('##################  SOAL 2  ##################')

print('hasil pengurangan:', np.subtract(a, b))
print('hasil pembagian:', np.divide(a, b))
```

```python
matrix_a = np.matrix(
 [[1, 2, 3],
 [4, 5, 6],
 [7, 8, 9]]
)
matrix_b = np.matrix(
 [[1, 4, 3],
 [2, 6, 6],
 [7, 0, 9]]
)
print(matrix_a)
```

```python
# Transpose
transposed_matrix = np.transpose(matrix_a)
print('transposed:')
print(transposed_matrix)
# Invers
inversed_matrix = np.linalg.inv(matrix_b)
print('inversed:')
print(inversed_matrix)
```

```python
# dot product
print('# dot product')
print(np.dot(matrix_a , matrix_b))

# penjumlahan
print("# penjumlahan")
print(np.add(matrix_a , matrix_b))

# pengurangan
print('# pengurangan')
print(np.subtract(matrix_a , matrix_b))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, fit_reg = False)
plt.show()
```

```python
# menggambarkan scatter plot sekaligus garis regresi
sns.lmplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
# scatterplot yang warnanya dibedakan berdasarkan suatu nilai, dibawah ini
# dibedakan dengan sex
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex',
fit_reg=False,
palette='Set1')
plt.show()
```

```python
tips['numb'] = range(1,245)
x = tips['numb']
y = tips['tip']
# normal plot
plt.plot(x,y)
# plot with data point marker
plt.plot(x,y, marker='o', color='purple', linewidth=0.1)
plt.show()
```

```python
sns.set_style("whitegrid")
tips = sns.load_dataset("tips")
sns.barplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
# Bar plot dipisah berdasarkan gender.
sns.barplot(x="day", y="total_bill", hue="sex", data=tips)
plt.show()
```

```python
##################  SOAL 5  ##################
print('##################  SOAL 5  ##################')

drinks_sorted = drinks.nlargest(5, 'beer_servings')
sns.barplot(x="country", y="beer_servings", data=drinks_sorted)
plt.show()
```
